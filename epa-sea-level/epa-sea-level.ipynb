```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import linregress

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df=pd.read_csv('epa-sea-level.csv')
```

```python
df.info()
```

```python
#clean the dataset containing nan
df = df.dropna(subset=['Year', 'CSIRO Adjusted Sea Level'])
```

```python
df
```

```python

df.plot(kind='scatter', x='Year', y='CSIRO Adjusted Sea Level', figsize=(6,6))
```

```python
linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
# Create scatter plot
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])



# Create the line of best fit
line1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'] )
x1 = np.arange(df['Year'].min(),2050,1)
y1 = x1*line1.slope + line1.intercept

slope = line1.slope
intercept = line1.intercept
# Print slope and intercept to verify
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

plt.plot(x1,y1, color='green')

# Add title and labels
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
    

# Save the plot as an image file
plt.savefig('sea_level_plot.png')

# Show the plot
plt.show()

# Predict the sea level in 2050
sea_level_2050 = intercept + slope * 2050
print(f"Predicted sea level in 2050: {sea_level_2050} inches")






 
```
